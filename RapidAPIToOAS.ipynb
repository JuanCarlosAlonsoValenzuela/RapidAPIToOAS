```python
def remove_tags(text):
    TAG_RE = re.compile(r'<[^>]+>')
    return TAG_RE.sub('', str(text)).strip()
```

```python
def obtain_html(url):
    hdr = {'User-Agent': 'Mozilla/5.0'}
    req = Request(url,headers=hdr)
    page = urlopen(req)
    html = page.read()
    bs = BeautifulSoup(html)
    return bs
```

```python
def obtain_data(url):
    hdr = {'User-Agent': 'Mozilla/5.0'}
    req = Request(url,headers=hdr)
    page = urlopen(req)
    html = page.read()
    bs = BeautifulSoup(html)
    return remove_tags(bs.find('script', {'id':'__NEXT_DATA__'}))
```

```python
def generate_paths(endpoints):
    print('paths')
    for path in endpoints:
        print('\t' + path['route'] + ':')
        
        print(2*'\t' + path['method'].lower() + ':')
        
        print(3*'\t' + 'summary: ' + path['summary'] ) if path['summary'] is not None else 0
        print(3*'\t' + 'description: ' + path['description']) if path['description'] is not None else 0
        print(3*'\t' + 'operationId: ' + path['id']) if path['id'] is not None else 0
        print(3*'\t' + 'parameters: ' )
        
        # Required parameteres
        print(4*'\t') 
        
```
